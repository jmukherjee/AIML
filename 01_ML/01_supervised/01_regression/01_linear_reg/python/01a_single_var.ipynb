# ML: Linear Regression - Single Variate
Linear Regression (Single Variate) is a statistical method where you use a linear regression model to predict a dependent variable based on only one independent variable, meaning you are analyzing the relationship between just two variables, making it a "single variate" analysis; also called "simple linear regression" in statistics

## Theory
- https://mlu-explain.github.io/linear-regression/

## Key points about *single variate* linear regression:
- **One predictor variable**: The key feature is that you only use one independent variable (X) to predict the dependent variable (Y). 
- **Straight line fit**: The model aims to find the best straight line that fits the data points, representing the linear relationship between X and Y. 
- **Equation**: The basic equation for single variate linear regression is typically written as: 
 `y = mx + b`, where
  - `m` is the slope of the line
  - `b` is the intercept
- **Examples**
  - Predicting house price based on square footage
  - Predicting weight based on height of a person

## Problem Statement

### 1.Predict House Price based on given House Area

| House Area | House Price |
| :--------: | :---------: |
|   sq.ft.   |      $.     |
|   2600.0   |  550000.00  |
|   3000.0   |  565000.00  |
|   3200.0   |  610000.00  |
|   3600.0   |  680000.00  |
|   4000.0   |  725000.00  |
Given above data build a machine learning model that can predict home prices based on given area in sq.ft.

```python
#pip install -r requirements.txt
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
data = {
    'area': [2600, 3000, 3200, 3600, 4000],
    'price': [550000, 565000, 610000, 680000, 725000]
}
test = {
    'area': [2000, 3300, 4100]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='magenta',marker='*')
```

```python
new_df = df.drop(['price'],axis=1)
new_df
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, df.price)
```

```python
tst = pd.DataFrame(test)
tst
```

```python
p=reg.predict(tst)
p
```

```python
tst['prices']=p
tst
```

```python
m = reg.coef_
m
```

```python
b = reg.intercept_
b
```

```python
plt.scatter(df.area,df.price,color='magenta',marker='*')
plt.plot(tst.area, tst.prices, color='blue')
```
