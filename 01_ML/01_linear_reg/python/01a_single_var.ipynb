# ML: Linear Regression - Single Variate
Linear Regression (Single Variate) is a statistical method where you use a linear regression model to predict a dependent variable based on only one independent variable, meaning you are analyzing the relationship between just two variables, making it a "single variate" analysis; also called "simple linear regression" in statistics

## Key points about *single variate* linear regression:
- **One predictor variable**: The key feature is that you only use one independent variable (X) to predict the dependent variable (Y). 
- **Straight line fit**: The model aims to find the best straight line that fits the data points, representing the linear relationship between X and Y. 
- **Equation**: The basic equation for single variate linear regression is typically written as: 
 `y = mx + b`, where
  - `m` is the slope of the line
  - `b` is the intercept
- **Examples**
  - Predicting house price based on square footage
  - Predicting weight based on height of a person
